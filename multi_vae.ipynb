```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, losses
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt

import sys
DATA_PATH = '/home/risitop/OneDrive/Documents/PHD/y1/data/EwingSarcoma_Miller2020/'
sys.path.append(DATA_PATH)

gpus = tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(gpus[0], True)
```

```python
# Sampling scheme
from tensorflow.keras import backend as K

def sampling(args):
    muA, logsA, muB, logsB, s = args
    zdim = muA.shape[1]
    epsilon = K.random_normal(shape=(K.shape(muA)[0], zdim),
                              mean=0., stddev=0.1)
    return s*muA + (1-s)*muB + K.exp(s*logsA + (1-s)*logsB) * epsilon

def selecting(args):
    xA, xB, s = args
    return [(1-s)*xA, s*xB]
```

```python
print("-- Loading and normalizing CHLA9...", end=' ')
XA = np.loadtxt(DATA_PATH + "CHLA9_pp.txt", delimiter=",").astype('float32')
XA = (XA - XA.min()) / (XA.max() - XA.min())
nA = len(XA)
print("Successfully loaded %i cells." %  nA)

print("-- Loading and normalizing CHLA10...", end=" ")
XB = np.loadtxt(DATA_PATH + "CHLA10_pp.txt", delimiter=",").astype('float32')
XB = (XB - XB.min()) / (XB.max() - XB.min())
nB = len(XB)
print("Successfully loaded %i cells." % nB)

XAB = np.concatenate( (XA, XB) )
s = np.concatenate( (np.zeros(nA), np.ones(nB) ) )
dA = XA.shape[1]
dB = XB.shape[1]
n = nA + nB
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)

xpcaA = pca.fit_transform(XA)
xpcaB = pca.fit_transform(XB)
xpcaAB = pca.fit_transform(XAB)

plt.figure(figsize=(20, 6))

plt.subplot(1,3,1)
plt.scatter(xpcaA[:, 0], xpcaA[:, 1], c=XA.sum(axis=1), ec='k', linewidth=0.5, s=50)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("CHLA9")

plt.subplot(1,3,2)
plt.scatter(xpcaB[:, 0], xpcaB[:, 1], c=XB.sum(axis=1), ec='k', linewidth=0.5, s=50)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("CHLA10")

plt.subplot(1,3,3)
plt.scatter(xpcaAB[:, 0], xpcaAB[:, 1], c=s, ec='k', linewidth=0.5, s=50)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("CHLA9 + CHLA10")
```

```python
 # [XA 0 ; 0 XB]
XAf = np.concatenate( (XA, np.zeros((nB, dA))) ) 
XBf = np.concatenate( (np.zeros((nA, dB)), XB) )
s = np.concatenate( (np.zeros(nA), np.ones(nB) ) ) # [0*nA 1*nB]

# Shuffling datasets
inds = np.arange(len(XAf))
np.random.shuffle(inds)

XAf = XAf[inds]
XBf = XBf[inds]
s = s[inds]
```

```python
ratio_test = 0.2
select = np.random.random((n,)) < ratio_test
nselect = np.logical_not(select)
XAf_train, XBf_train, s_train = XAf[nselect], XBf[nselect], s[nselect]
XAf_test, XBf_test, s_test = XAf[select], XBf[select], s[select]

intermediate_dim = 128
latent_dim = 32

print("Train size:", XAf_train.shape[0])
print("Test size:", XAf_test.shape[0])
print("Input shape A:", dA)
print("Input shape B:", dB)
print("Intermediate shape:", intermediate_dim)
print("Latent dimension:", latent_dim)
```

```python
## Building the model

# Encoding layers, space A
iA = keras.Input(shape=(dA,))
hA = layers.Dense(intermediate_dim, activation='relu')(iA)
muA = layers.Dense(latent_dim)(hA)
logsA = layers.Dense(latent_dim)(hA)

# Encoding layers, space B
iB = keras.Input(shape=(dB,))
hB = layers.Dense(intermediate_dim, activation='relu')(iB)
muB = layers.Dense(latent_dim)(hB)
logsB = layers.Dense(latent_dim)(hB)

# Original dataset indication
s = keras.Input(shape=(1,))

# Common latent space
z = layers.Lambda(sampling)([muA, logsA, muB, logsB, s])

# Decoder
zi = keras.Input(shape=(latent_dim,), name='z_sampling')
gA = layers.Dense(intermediate_dim, activation='relu')(zi)
gB = layers.Dense(intermediate_dim, activation='relu')(zi)
oA = layers.Dense(dA, activation='sigmoid')(gA)
oB = layers.Dense(dB, activation='sigmoid')(gB)

encoder = keras.Model([iA, iB, s], [muA + muB, logsA + logsB, z], name='encoder')
decoder = keras.Model(zi, [oA, oB], name='decoder')
```

```python
# instantiate VAE model
outputs = selecting(decoder(encoder([iA, iB, s])[2]) + [s])
vae = keras.Model([iA, iB, s], outputs, name='vae_mlp')
```

```python
reconstruction_loss = keras.losses.binary_crossentropy(
    [iA, iB], 
    selecting(decoder(encoder(
        selecting(decoder(encoder([iA, iB, s])[2]) + [1 - s]) + [1 - s]
    )[2]) + [s])
)
reconstruction_loss *= dA # To adjust in the future

kl_loss = 2 + logsA + logsB - K.square(muA) - K.square(muB) - K.exp(logsA) - K.exp(logsB) # To recompute
kl_loss = K.sum(kl_loss, axis=-1)
kl_loss *= -0.25

vae_loss = K.mean(reconstruction_loss + kl_loss)
vae.add_loss(vae_loss)
opt = keras.optimizers.Adam(learning_rate=1e-4)
vae.compile(optimizer=opt)
```

```python
vae.fit([XAf_train, XBf_train, s_train], [XAf_train, XBf_train],
        epochs=20,
        batch_size=32,
        validation_data=([XAf_train, XBf_train, s_train], [XAf_train, XBf_train]))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

x_test_encoded = encoder.predict([XAf_train, XBf_train, s_train], batch_size=32)[2]
x_pca = pca.fit_transform(x_test_encoded)
x_pca = x_test_encoded

plt.figure(figsize=(6, 6))

plt.scatter(x_pca[:, 0], x_pca[:, 1], c=s_train, ec='k', linewidth=0.5, s=50)
plt.xlabel("D1 (latent space)")
plt.ylabel("D2 (latent space)")
plt.title("CHLA9, VAE latent space")
```
